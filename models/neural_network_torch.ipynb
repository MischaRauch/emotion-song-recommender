```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics

import torch
import torch.nn as nn
```

```python
df = pd.read_csv('../data_moods.csv')
df
```

```python
moods= {'Sad' : 0, 'Calm' : 1, 'Energetic' : 2, 'Happy' : 3}
df['mood'] = df['mood'].apply(lambda x: moods[x])
df['mood'].value_counts()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
#df['release_year'] = df['release_date'].dt.year
#df['release_month']= df['release_date'].dt.month
#df['release_day']= df['release_date'].dt.day

X = df.drop(['mood','id','name','album','artist','release_date'],axis=1)
y = df['mood']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.09, random_state=42)
X
```

```python
# Create Datasets
x_train = torch.tensor(X_train.values, dtype=torch.float)
y_train = torch.tensor(y_train.values, dtype=torch.int64)
train_ds = torch.utils.data.TensorDataset(x_train, y_train)

x_test = torch.tensor(X_test.values, dtype=torch.float)
y_test = torch.tensor(y_test.values, dtype=torch.int64)
test_ds = torch.utils.data.TensorDataset(x_test, y_test)
```

```python
# Create Dataloaders
train_dl = torch.utils.data.DataLoader(train_ds, batch_size=25, shuffle=True)
test_dl = torch.utils.data.DataLoader(test_ds, batch_size=25, shuffle=False)
```

```python
class MyANN(nn.Module):
  
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(13, 32)  
    self.av1 = nn.ReLU()
    self.dp1 = nn.Dropout(0.5)
    self.fc2 = nn.Linear(32, 64)  
    self.av2 = nn.ReLU()
    self.dp2 = nn.Dropout(0.5)
    self.fc3 = nn.Linear(64, 32)  
    self.av3 = nn.ReLU()
    self.dp3 = nn.Dropout(0.5)
    self.fc4 = nn.Linear(32, 4)  
    self.avout = nn.LogSoftmax(dim=-1)
    return


  def forward(self, x):
    x = self.fc1(x)
    x = self.av1(x)
    x = self.dp1(x)
    x = self.fc2(x)
    x = self.av2(x)
    x = self.dp2(x)
    x = self.fc3(x)
    x = self.av3(x)
    x = self.dp3(x)
    x = self.fc4(x)
    x = self.avout(x)      # Alternate way: nn.functional.log_softmax(x)
    return x
```

```python
class MyANN2(nn.Module):
  
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(13, 20)  
    self.av1 = nn.ReLU()
    self.dp1 = nn.Dropout(0.5)
    self.fc2 = nn.Linear(20, 20)  
    self.av2 = nn.ReLU()
    self.dp2 = nn.Dropout(0.5)
    self.fc3 = nn.Linear(20, 13)  
    self.av3 = nn.ReLU()
    self.dp3 = nn.Dropout(0.5)
    self.fc4 = nn.Linear(13, 4)  
    self.avout = nn.LogSoftmax(dim=-1)
    return


  def forward(self, x):
    x = self.fc1(x)
    x = self.av1(x)
    x = self.dp1(x)
    x = self.fc2(x)
    x = self.av2(x)
    x = self.dp2(x)
    x = self.fc3(x)
    x = self.av3(x)
    x = self.dp3(x)
    x = self.fc4(x)
    x = self.avout(x)      # Alternate way: nn.functional.log_softmax(x)
    return x
```

```python
class MyANN3(nn.Module):
  
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(13, 8)  
    self.av1 = nn.ReLU()
    self.dp1 = nn.Dropout(0.5)
    self.fc3 = nn.Linear(8, 4)  
    self.avout = nn.LogSoftmax(dim=-1)
    return


  def forward(self, x):
    x = self.fc1(x)
    x = self.av1(x)
    x = self.dp1(x)
    x = self.fc3(x)
    x = self.avout(x)      # Alternate way: nn.functional.log_softmax(x)
    return x
```

```python
model = MyANN3()
print(model)
```

```python
# Train my neural network.
optimizer = torch.optim.Adam(model.parameters(), lr=0.0006)
for epochs in range(50):
  for data in train_dl:
    X, y = data
     # print(data)
    optimizer.zero_grad() # clear gradient information.
    output = model(X.view(-1, 13))
    loss = nn.functional.nll_loss(output, y)
    loss.backward() # do pack-propagation step
    optimizer.step() # update weights.
  print(loss.data)
```

```python
# Evaluate the trained network.
total = 0
correct = 0
with torch.no_grad():   # No need for keepnig track of necessary changes to the gradient.
  for data in test_dl:
    X, y = data
    output = model(X.view(-1, 13))
    for idx, val in enumerate(output):
      if torch.argmax(val) == y[idx]:
        correct += 1
      total += 1
  print('Accuracy:', round(correct/total, 3))
```
