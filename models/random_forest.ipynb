```python
# https://github.com/cristobalvch/Spotify-Machine-Learning

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
df = pd.read_csv('../data_moods.csv')
df
```

```python
moods= {'Sad' : 0, 'Calm' : 1, 'Energetic' : 2, 'Happy' : 3}
df['mood'] = df['mood'].apply(lambda x: moods[x])
df['mood'].value_counts()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
#df['release_year'] = df['release_date'].dt.year
#df['release_month']= df['release_date'].dt.month
#df['release_day']= df['release_date'].dt.day

X = df.drop(['mood','id','name','album','artist','release_date'],axis=1)
y = df['mood']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X
```

```python

forest =  RandomForestClassifier(bootstrap=True, max_features=5, n_estimators=10, random_state=0)

trained_forest = forest.fit(X_train, y_train)

y_pred_train = trained_forest.predict(X_train)
print('Accuracy on training data= ', metrics.accuracy_score(y_true = y_train, y_pred = y_pred_train))

y_pred = trained_forest.predict(X_test)
print('Accuracy on test data= ', metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
```
